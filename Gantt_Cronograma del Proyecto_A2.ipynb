```python
import matplotlib.pyplot as plt
import numpy as np

# Define tasks and their durations in weeks
steps = [
    "Análisis de Mercado y Necesidades",
    "Recolección de Datos",
    "Diseño del Sistema",
    "Procesamiento y Análisis de Datos",
    "Desarrollo del Sistema de Recomendación",
    "Pruebas y Optimización",
    "Despliegue y Lanzamiento",
    "Promoción y Mantenimiento"
]
updated_durations = [1, 1, 1, 1, 3, 1, 1, 1]

# Calculate total weeks and days
total_weeks = sum(updated_durations)
total_days = total_weeks * 7  # 7 days per week

# Create a transparent grid for the tasks
task_grid = np.zeros((len(steps), total_days, 4))  # RGBA format

import matplotlib.pyplot as plt
import numpy as np

# Define custom colors for GB
custom_colors = [
    (144, 202, 2490),   
    (129, 212, 250),   
    (128, 222, 234),    
    (128, 203, 196),  
    (165, 214, 167),   
    (197, 225, 165),  
    (230, 238, 156), 
    (255, 245, 157)    
]

# Convert custom colors to the range [0, 1] and create an RGBA array
task_colors = np.array(custom_colors) / 255.0

# Define tasks and their durations in weeks
steps = [
    "Análisis de Mercado y Necesidades",
    "Recolección de Datos",
    "Diseño del Sistema",
    "Procesamiento y Análisis de Datos",
    "Desarrollo del Sistema de Recomendación",
    "Pruebas y Optimización",
    "Despliegue y Lanzamiento",
    "Promoción y Mantenimiento"
]
updated_durations = [1, 1, 1, 1, 3, 1, 1, 1]

# Calculate total weeks and days
total_weeks = sum(updated_durations)
total_days = total_weeks * 7  # 7 days per week

# ransparent grid for the tasks
task_grid = np.zeros((len(steps), total_days, 4))  # RGBA format

# Reset current_day to 0 for proper indexing
current_day = 0

for i, duration in enumerate(updated_durations):
    for day in range(duration * 7):
        if day % 7 < 5:  # Only mark weekdays
            task_grid[i, current_day + day, :3] = task_colors[i]  # RGB color
            task_grid[i, current_day + day, 3] = 1  # Alpha (opacity)
    current_day += duration * 7
    
# Plotting
plt.figure(figsize=(14, 8))

# Plot the task grid
plt.imshow(task_grid, aspect='auto')

# lines to represent the grid for each day of the week
for i in range(1, total_days):
    plt.axvline(x=i, color='grey', linestyle='-', linewidth=0.5)

for i in range(1, len(steps)):
    plt.axhline(y=i, color='grey', linestyle='-', linewidth=0.5)

# Set labels
plt.yticks(ticks=np.arange(len(steps)), labels=steps)
plt.xticks(ticks=np.arange(0, total_days, 7), labels=[f"S{i+1}" for i in range(total_weeks)])

plt.xlabel('Días (S1, S2, S3, ...)')
plt.title('Plan de Proyecto: Sistema de Recomendación de Restaurantes')

plt.show()
```
